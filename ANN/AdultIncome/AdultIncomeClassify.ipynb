#### Create tf.feature_columns for categorical values.Use vocabulary lists or just use hash buckets

```python
cont_feat = [tf.feature_column.numeric_column(k) for k in num_col]
cont_feat
```

```python
relationship = train['relationship'].value_counts().index.to_list()
relationship
```

```python
relationship = tf.feature_column.categorical_column_with_vocabulary_list('relationship' , relationship)
relationship
```

```python
cat_feat = [tf.feature_column.categorical_column_with_hash_bucket(k,hash_bucket_size = 1000) for k in cat_col]
cat_feat
```

```python
model = tf.estimator.LinearClassifier(n_classes = 2 , model_dir = r'ongoing\model', 
                                      feature_columns = cat_feat + cont_feat)
```

```python
features = train.columns.to_list()
label = 'label'
```

```python
features.remove('label')
```

```python
features
```

```python
def get_input_fn(data_set , num_epochs = None , n_batch = 128 , shuffle = True):
    return tf.compat.v1.estimator.inputs.pandas_input_fn(
        x = pd.DataFrame({k: data_set[k].values for k in features}),
        y = pd.Series(data_set[label].values),
        batch_size = n_batch ,
        num_epochs = num_epochs ,
        shuffle = shuffle)
```

```python
model.train(input_fn = get_input_fn(train , num_epochs = None , n_batch = 128 , shuffle = False) , steps = 1000)
```

```python
model.evaluate(input_fn = get_input_fn(test , num_epochs = 1 , n_batch = 128 , shuffle = False) , steps = 1000)
```

```python
# salary is 0 at a very young age & then keeps increasing . Close to retirement it again decreases. hence we square it

def square_var(train , test , var = 'age'):
    train['new'] = train[var].pow(2)
    test['new'] = test[var].pow(2)    
    
    return train , test
```

```python
train , test = square_var(train , test , var = 'age')
```
